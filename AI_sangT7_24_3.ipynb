```python
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
print(x[:5])
print(y[:5])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
sc=StandardScaler()
sc.fit(x_train)
x_train_Std=sc.transform(x_train)
x_test_Std=sc.transform(x_test)
```

```python
print(x_train,x_train_Std)
```

```python
model=Perceptron(max_iter=10000, eta0=0.05,random_state=0)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(y_pred)
print(y_test)
print('Độ chính xác của mô hình chưa Scale:',accuracy_score(y_test,y_pred))
```

```python
model=Perceptron(max_iter=10000, eta0=0.05,random_state=0)
model.fit(x_train_Std,y_train)
y_pred = model.predict(x_test_Std)
print(y_pred)
print(y_test)
print("Độ chính xác của mô hình đã Scale:",accuracy_score(y_test,y_pred))
```
