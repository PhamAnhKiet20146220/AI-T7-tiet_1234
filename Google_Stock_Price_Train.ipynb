```python
import pandas as pd
url = '/content/Google_Stock_Price_Train.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot= True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['High']), y = data['Open'])
print('He so intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['High']), y = data['Open'])
```

```python
y_pred = model.predict(X=pd.DataFrame(data['High']))
```

```python
data.plot(kind='scatter',x='High', y = 'Open' , figsize=(9,9),
          color= 'black')
plt.plot(data['High'],y_pred,color='blue')
```
